# Venn 4 sets

```python
from venny4py.venny4py import *
%matplotlib inline

#dict of sets
sets = {
    'Set1': set(list("Harry Potter")),
    'Set2': set(list("Hermione Granger")),
    'Set3': set(list("Ron Weasley")),
    'Set4': set(list("Severus Snape"))
}

venny4py(sets=sets, out='Example_output')
```

# Specify dpi and figsize

```python
from venny4py.venny4py import *
%matplotlib inline

#dict of sets
sets = {
    'Set1': set(list("Harry Potter")),
    'Set2': set(list("Hermione Granger")),
    'Set3': set(list("Ron Weasley")),
    'Set4': set(list("Severus Snape"))
}

venny4py(sets=sets, size=6, dpi=600, out='Example_output', ext='svg')
```

# As ax, or matplotlib subplot

```python
from venny4py.venny4py import *
import matplotlib.pyplot as plt
%matplotlib inline

#dict of sets
sets = {
    'Set1': set(list("Harry Potter")),
    'Set2': set(list("Hermione Granger")),
    'Set3': set(list("Ron Weasley")),
    'Set4': set(list("Severus Snape"))
}

fig, axes = plt.subplots(1, 2, figsize=(6, 3), dpi=600)
ax = axes[1]
venny4py(sets=sets, size=3, asax=ax) #indicate approx. size of subplot, not the figure!
```
